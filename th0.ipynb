```python
#new
df=pd.read_csv("/content/boston_data.csv")
df.head()
```

```python
plt.boxplot(df['crim'])
```

```python
Q3=df["crim"].quantile(0.75)
Q1=df["crim"].quantile(0.25)
IQR=Q3-Q1
lower_bound=Q1-(1.5*IQR)
upper_bound=Q3+(1.5*IQR)
outliers_data=np.where(df["crim"]>upper_bound,True,np.where(df["crim"]<lower_bound,True,False))
df[outliers_data]
trimmed_data=df[~outliers_data]
print(lower_bound)
print(upper_bound)
plt.boxplot(trimmed_data['crim'])
plt.title("trimmed crim")
plt.show()
```

```python
plt.boxplot(df['dis'])
```

```python
Q3=df["dis"].quantile(0.75)
Q1=df["dis"].quantile(0.25)
IQR=Q3-Q1
lower_bound=Q1-(1.5*IQR)
upper_bound=Q3+(1.5*IQR)
outliers_data=np.where(df["dis"]>upper_bound,True,np.where(df["dis"]<lower_bound,True,False))
df[outliers_data]
trimmed_data=df[~outliers_data]
print(lower_bound)
print(upper_bound)
plt.boxplot(trimmed_data['dis'])
plt.title("trimmed dis")
plt.show()
```
