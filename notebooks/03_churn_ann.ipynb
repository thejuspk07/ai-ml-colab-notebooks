```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
data = pd.read_csv(r"/content/Churn-Modelling (3).csv")
```

```python
X= data.iloc[:, 3:-1].values   # exclude RowNumber, CustomerId, Surname
y = data.iloc[:, -1].values     # target (Exited)
```

```python
data
```

```python
le = LabelEncoder()
X[:, 2] = le.fit_transform(X[:, 2])
```

```python
ct = ColumnTransformer(transformers=[("encoder", OneHotEncoder(), [1])], remainder="passthrough")
X = np.array(ct.fit_transform(X))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
X.shape
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
ann = tf.keras.models.Sequential()

# Input + Hidden layers
ann.add(tf.keras.layers.Dense(units=6, activation="relu"))
ann.add(tf.keras.layers.Dense(units=6, activation="relu"))
```

```python
ann.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
```

```python
ann.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
history = ann.fit(X_train, y_train, batch_size=32, epochs=100, verbose=1, validation_data=(X_test, y_test))
```

```python
y_pred = ann.predict(X_test)
```

```python
y_pred_binary = (y_pred > 0.5) # Convert probabilities to binary predictions
cm = confusion_matrix(y_test, y_pred_binary)
print("\nConfusion Matrix:\n", cm)
```

```python
acc = accuracy_score(y_test, y_pred_binary)
print("\nTest Accuracy: {:.2f}%".format(acc * 100))
```

```python
print("\nClassification Report:\n", classification_report(y_test, y_pred_binary))
```
