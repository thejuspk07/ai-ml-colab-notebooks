```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
cancer = load_breast_cancer()
cancer.keys()
```

```python
df = pd.DataFrame(cancer.data,columns=cancer.feature_names)
df.head()
```

```python
scaler = StandardScaler()
scaler.fit(df)
```

```python
scaled_features = scaler.transform(df)
pca =PCA(n_components=3)
pca.fit(scaled_features)
```

```python
x_pca =scaler.transform(scaled_features)
plt.figure(figsize=(8,6))
plt.xlabel("first principle component")
plt.ylabel("second principle component")
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'],cmap='plasma',s=50)
plt.show()
```

```python
df_comp = pd.DataFrame(pca.components_,columns=cancer.feature_names)
sns.heatmap(df_comp,cmap='plasma')
```

```python
pca.fit(scaled_features)
x_pca=scaler.transform(scaled_features)
fig = plt.figure(figsize=(10,10))
ax=fig.add_subplot(111,projection='3d')
ax.set_xlabel("first component axis")
ax.set_ylabel("secondcomponent axis")
ax.set_zlabel("third component axis")
ax.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'],cmap='plasma',s=50)
df.corr(method='pearson')

```
