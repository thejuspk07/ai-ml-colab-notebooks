# Simple Linear Regression

## Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing the dataset

```python
df=pd.read_csv("/content/Salary.csv")
df.head()
```

## Splitting the dataset into the X and Y

```python

x=df[["YearsExperience"]]
y=df[["Salary"]]
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y,color="red")
plt.title("Salary vs Experience")
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.show()
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
#0.2=split 20% for test
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

## Predicting the Test set results

```python
y_pred_test=model.predict(x_test) # for test
y_pred_train=model.predict(x_train) # for train
```

```python
#predicted for test
y_pred_test
```

```python
# actual for test
y_test
```

```python
#predicted for train
y_pred_train
```

```python
#actual  for train
y_train
```

## Visualising the Training set results

```python
# Salary vs Experience (Training Set)
plt.scatter(x_train, y_train, color="red")
plt.plot(x_train, y_pred_train, color="blue")
plt.title("Salary vs Experience (Training Set)")
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.show()
```

## Visualising the Test set results

```python
# Salary vs Experience (Test Set)
plt.scatter(x_test,y_test,color="red")
plt.plot(x_test,y_pred_test,color="blue")
plt.title("Salary vs Experience (Test Set)")
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
```

## Let's predict a salary of an employee with number of years of experience

```python
new=float(input("Enter the years of experience:"))
result=model.predict([[new]])
print(result)
```

## Error Calculation

```python
#mse
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,y_pred_test)
mse
```

```python
#rmse
from sklearn.metrics import mean_squared_error
rmse=np.sqrt(mean_squared_error(y_test,y_pred_test))
rmse
```

```python
#mae
mae=mean_squared_error(y_test,y_pred_test)
mae
```

## Model performance

```python
# Accuracy
```

```python
# for train
from sklearn.metrics import r2_score
acc_train=r2_score(y_train,y_pred_train)
acc_train
```

```python
#another method for accuracy
from sklearn.metrics import r2_score
acc_test = r2_score(y_test, y_pred_test)
acc_test
```

```python
# for test
from sklearn.metrics import r2_score
acc_test=r2_score(y_test,y_pred_test)
acc_test
```

## Model Performance Summary

Based on the calculated metrics:

- Mean Squared Error (MSE): {{mse}}
- Root Mean Squared Error (RMSE): {{rmse}}
- Mean Absolute Error (MAE): {{mae}}
- R-squared (Training Set): {{acc_train}}
- R-squared (Test Set): {{acc_test}}

These values indicate how well your linear regression model fits the data and its predictive capability. A lower MSE, RMSE, and MAE indicate less error, while an R-squared value closer to 1 indicates a better fit.

```python
#Get number of samples and features
n_test=len(y_test)
n_train=len(y_train)
p=x_train.shape[1]#Assuming x train is your feature matrix
```

```python
#Adjusted R square
adjusted_r2_test=1-(1-acc_test)*(n_test-1)/(n_test-p-1)
adjusted_r2_train=1-(1-acc_train)*(n_train-1)/(n_train-p-1)

```

```python
#print
print("Adjusted R-squared (Test Set):", adjusted_r2_test)
print("Adjusted R-squared (Train Set):", adjusted_r2_train)
```
